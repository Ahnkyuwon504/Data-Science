```python
import pandas as pd
import io
from google.colab import files

# uploaded = files.upload()
# df = pd.read_csv(io.StringIO(uploaded['mtcars.csv'].decode('utf-8')))

df = pd.read_csv("/content/drive/MyDrive/빅데이터예측분석/Life Expectancy Data_1.csv", sep=r'\s*,\s*')

# df 처리
df.fillna(df.mean(), inplace=True)
df.dropna()
df.head()

```

```python
from sklearn.model_selection import train_test_split

# sklearn package
x = df[['Status', 'Adult Mortality', 'infant deaths', 'Alcohol']]

y = df[['expectancy']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)

# N/A 처리
x_test.fillna(x_test.mean())
```

```python
from sklearn.linear_model import LinearRegression

# 모델생성
mlr = LinearRegression()
mlr.fit(x_train.values, y_train.values) # values
```

```python
# 임의의 값을 통해 mpg predict
my_vars = [[4, 70, 80, 3, 2, 20, 1, 1, 4, 1]]
my_predict = mlr.predict(my_vars)
print(my_predict)
```

```python
# test데이터 넣어 예측 .. y_test와 비교를위함
y_predict = mlr.predict(x_test.values)
```

```python
# 시각화
import matplotlib.pyplot as plt

plt.scatter(y_test, y_predict, alpha=0.4)
plt.xlabel("Actual mpg")
plt.ylabel("Predicted mpg")
plt.title("MULTIPLE LINEAR REGRESSION")
plt.show()
```

```python
# weight 확인
print(mlr.coef_)

# 결정계수
print(mlr.score(x_train.values, y_train.values))

```
